```python
def remove_punctuation(text):
    punctuation_pattern = re.compile(r'[^\w\s]')
    return re.sub(punctuation_pattern, '', text)
```

```python
data["Verse"]=data["Verse"].apply(remove_punctuation)
data.head()
```

```python
import nltk
nltk.download('punkt')
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(text):
    stop_words = set(stopwords.words("english"))
    words = nltk.word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)
     
```

```python
data["Verse"]=data["Verse"].apply(remove_stop_words)
data.head()
```

```python
text_list = data['Verse'].tolist()
print(text_list)
```

```python
import nltk
nltk.download('punkt')

data = text_list

def process_row(row):
    tokens = nltk.word_tokenize(row)
    return tokens

processed_data = [process_row(row) for row in data]

processed_data
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(processed_data)
```

```python
model.train(processed_data, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
# Tokenize the "Verse" column into sentences
sentences = [nltk.sent_tokenize(verse) for verse in text_list]

# Flatten the list of lists into a single list
sentences = [sentence for sublist in sentences for sentence in sublist]

# Tokenize each sentence into words
words = [nltk.word_tokenize(sentence) for sentence in sentences]

# Flatten the list of lists into a single list
words = [word for sublist in words for word in sublist]
```
